## Hypothesis Testing

Let's take a look at one of the studies in our dataset: the 'Nac/NtrC' Study.

```python
nac_sample_table = sample_table.loc[sample_table['Study'] == 'Nac/NtrC']
nac_sample_table
```

Here, two main things vary: the nitrogen source, and the strain description. The two strains are 'MG1655' and 'BW25113 del_nac'. The two base strains are closely related, but the BW25113 samples in this project have the transcription factor, Nac, deleted. Nac is involved in the response to nitrogen starvation.

This poses the question: **Which genes are significantly affected by the genetic differences between these two strains, regardless of the nitrogen source used?** The answer to this suggests which genes Nac controls.

### Generate p-values for each gene

For each gene, compare the expression in the Nac/NtrC BW25113 del_nac samples with the Nac/NtrC MG1655 samples. Before doing so, think carefully about your approach:

- Given the sample size here, can you use a non-parametric test?
- If you choose to use a test that assumes a distribution, you must also generate a probability (p-value) for your assumptions. ([Hint](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html))

For all p-values, also generate a corresponding boolean (true/false) value. For assumptions, this value should be True if the assumption passes. For the comparison p-value, this value should be True if the gene is significantly differently expressed. Use $\alpha$ = 0.05.

#### Discuss your choice of tests

<span style="color:red">5 points</span>

Discuss the choice of tests:
there are ~14 samples and 3887 thousand genes. It's hard to tell whether I should be using a parametric test or non-parametric test alone. 
Based on the hint provided, my logic here is:
1) for each gene, perform a shapiro-wilk test on both samples (BW25113 and MG1655)

2) if the shapiro-wilk test tells me that both samples can be assumed normal, then I can use a parametric test (t-test) to see if they come from the same distribution

3) if only one or neither samples can be assumed normal, then I can need to use a non-parametric test (wilcoxon test) to see if they come from the same distribution. Since the two samples have different length, I am using a Mann-Whitney test instead.

#### Generate your p-values

Use a table or visualization of your significance results so we can get a sense of how many genes are significant and whether the assumptions pass. Comment on your results.

<span style="color:red">10 points</span>

```python
# perform a Shapiro-Wilk test for normality.
alpha = 0.05 # specified by the question

# select the samples' data
names = nac_sample_table.index.tolist()
samples = data[names].copy()
samples_1 = data[['nac_ntrc__wt_csn__1','nac_ntrc__wt_csn__2','nac_ntrc__wt_cytd__1','nac_ntrc__wt_cytd__2','nac_ntrc__wt_gln__1','nac_ntrc__wt_gln__2']].copy()
samples_2 = data[['nac_ntrc__bw_delnac_csn__1','nac_ntrc__bw_delnac_csn__2','nac_ntrc__bw_delnac_cytd__1','nac_ntrc__bw_delnac_cytd__2','nac_ntrc__bw_delnac_gln__1','nac_ntrc__bw_delnac_gln__2','nac_ntrc__bw_delnac_nh4__1','nac_ntrc__bw_delnac_nh4__2']].copy()
```

```python
from scipy.stats import mannwhitneyu
from scipy.stats import ttest_ind

# create place holders
assumption_p_1 = []
assumption_p_b_1 = []
assumption_p_2 = []
assumption_p_b_2 = []
assumption_boolean = []
test_p = []
test_boolean = []

# perform SW test for each gene
for i in samples.index:
    
    # sample 1
    shapiro_test_1 = stats.shapiro(samples_1.loc[i])
    assumption_p_1.append(shapiro_test_1.pvalue)
    # if p is smaller or equal to alpha, reject null hypothesis. -> not from a normal distribution
    if shapiro_test_1.pvalue <= alpha:
        assumption_p_b_1.append(False)
    # if p is larger than alpha, do not reject null hypothesis. -> from a normal distribution    
    else:
        assumption_p_b_1.append(True)
    
    # sample 2
    shapiro_test_2 = stats.shapiro(samples_2.loc[i])
    assumption_p_2.append(shapiro_test_2.pvalue)
    # if p is smaller or equal to alpha, reject null hypothesis. -> not from a normal distribution
    if shapiro_test_2.pvalue <= alpha:
        assumption_p_b_2.append(False)
    # if p is larger than alpha, do not reject null hypothesis. -> from a normal distribution    
    else:
        assumption_p_b_2.append(True)
    
    # assumption boolean
    if shapiro_test_1.pvalue <= alpha or shapiro_test_2.pvalue <= alpha:
        assumption_boolean.append(False) # not from normal distribution need to use a mann-whitney test
        test_p_value = mannwhitneyu(samples_1.loc[i],samples_2.loc[i])[1]
        test_p.append(test_p_value)
        if test_p_value <= alpha:
            test_boolean.append(False) # -> not from the same distribution
        else:
            test_boolean.append(True) # -> from the same distribution    
    else:
        assumption_boolean.append(True) # both from normal distribution, can use a t-test
        test_p_value = ttest_ind(samples_1.loc[i],samples_2.loc[i])[1]
        test_p.append(test_p_value)
        if test_p_value <= alpha:
            test_boolean.append(False) # -> not from the same distribution
        else:
            test_boolean.append(True) # -> from the same distribution
        

samples['SW P-value MG1655'] = assumption_p_1
samples['SW P-value Boolean MG1655'] = assumption_p_b_1
samples['SW P-value BW25113'] = assumption_p_2
samples['SW P-value Boolean BW25113'] = assumption_p_b_2
samples['SW P-value Assumption Boolean'] = assumption_boolean
samples['Comparison P-value'] = test_p
samples['Comparison P-value Boolean'] = test_boolean
samples
```

```python
#visualize the result on the assumption
samples['SW P-value MG1655'].hist()
plt.title('Assumption P-value results')
plt.xlabel('P-value')
plt.ylabel('count')

samples['SW P-value BW25113'].hist()
plt.title('Assumption P-value results')
plt.xlabel('P-value')
plt.ylabel('count')
```

```python
# visualize the result on comparison
samples['Comparison P-value'].hist()
plt.title('Comparison P-value results')
plt.xlabel('P-value')
plt.ylabel('count')
```

```python
# first use FWER for the assumption test for both samples
bonferroni_1 = 0.05/len(samples_1)
bonferroni_2 = 0.05/len(samples_2)

assumption_p_1 = []
assumption_p_b_1 = []
assumption_p_2 = []
assumption_p_b_2 = []
assumption_boolean = []
test_p = []
test_boolean = []

# perform SW test for each gene after correction
for i in samples.index:
    
    # sample 1
    shapiro_test_1 = stats.shapiro(samples_1.loc[i])
    assumption_p_1.append(shapiro_test_1.pvalue)
    # if p is smaller or equal to bonferroni value, reject null hypothesis. -> not from a normal distribution
    if shapiro_test_1.pvalue <= bonferroni_1:
        assumption_p_b_1.append(False)
    # if p is larger than alpha, do not reject null hypothesis. -> from a normal distribution    
    else:
        assumption_p_b_1.append(True)
    
    # sample 2
    shapiro_test_2 = stats.shapiro(samples_2.loc[i])
    assumption_p_2.append(shapiro_test_2.pvalue)
    # if p is smaller or equal to bonferroni value, reject null hypothesis. -> not from a normal distribution
    if shapiro_test_2.pvalue <= bonferroni_2:
        assumption_p_b_2.append(False)
    # if p is larger than alpha, do not reject null hypothesis. -> from a normal distribution    
    else:
        assumption_p_b_2.append(True)
    
    # for the comparison part, I am having trouble implementing the FDR algorithm with ranks
    # sorry :(
    
    # assumption boolean
    if shapiro_test_1.pvalue <= bonferroni_1 or shapiro_test_2.pvalue <= bonferroni_2:
        assumption_boolean.append(False) # not from normal distribution need to use a mann-whitney test
    #    
    #    test_p_value = mannwhitneyu(samples_1.loc[i],samples_2.loc[i])[1]
    #    test_p.append(test_p_value)
    #    if test_p_value <= alpha:
    #        test_boolean.append(False) # -> not from the same distribution
    #    else:
    #        test_boolean.append(True) # -> from the same distribution    
    else:
        assumption_boolean.append(True) # both from normal distribution, can use a t-test
    #    
    #    test_p_value = ttest_ind(samples_1.loc[i],samples_2.loc[i])[1]
    #    test_p.append(test_p_value)
    #    if test_p_value <= alpha:
    #        test_boolean.append(False) # -> not from the same distribution
    #    else:
    #        test_boolean.append(True) # -> from the same distribution

samples['FWER SW P-value MG1655'] = assumption_p_1
samples['FWER SW P-value Boolean MG1655'] = assumption_p_b_1
samples['FWER SW P-value BW25113'] = assumption_p_2
samples['FWER SW P-value Boolean BW25113'] = assumption_p_b_2
samples['FWER SW P-value Assumption Boolean'] = assumption_boolean
samples
```
