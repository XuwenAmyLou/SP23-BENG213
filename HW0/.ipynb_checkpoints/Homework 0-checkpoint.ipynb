### Other Groupings: Strings, Sets, and Tuples

[Strings](https://www.w3schools.com/python/python_strings.asp) are essentially lists of characters used for names and words. There are many useful built-in string functions for operations like capitalization and searching for substrings.

[Sets](https://docs.python.org/3/tutorial/datastructures.html#sets) are unordered lists that can't contain repeat elements. They are the most useful when we want to perform set operations like union and intersection. For example, you might have two sets of important genes, and want to know [which genes appear in both sets](https://www.w3schools.com/python/ref_set_intersection.asp). You can convert between sets and lists using their respective constructors (set() and list()).

[Tuples](https://docs.python.org/3/library/stdtypes.html#tuples) are the least commonly used sequence types. They use parentheses instead of brackets and cannot be modified the way that lists can. Don't worry about them too much, but be aware that they can be converted back to lists using the list() constructor.

For each of these, know how they are constructed and what they can be used for. When you run into an application for them in a problem or you find them as an output to a function, you can always pull up the documentation for them.

### Other Useful Functions to Know

- [type()](https://www.programiz.com/python-programming/methods/built-in/type): Returns the type of a variable, which is useful for debugging and understanding code.
- [isinstance()](https://www.programiz.com/python-programming/methods/built-in/isinstance): If you want to do something different to a variable depending on its type, you could create an if statement using this function. For example, you might have a list that combines numbers and strings, and you want to round all of the numbers to a specific number of decimals but do nothing to the strings.
- [zip()](https://www.geeksforgeeks.org/zip-in-python/#highlighter_116733): If you are iterating through two lists at once, you can zip them together, which creates a tuple-like object.
- [enumerate()](https://www.geeksforgeeks.org/enumerate-in-python/): Similarly, if you are iterating through a list and also want the number of iterations, you can use this function. This is the same as zip(range(len(my_list)), my_list).

### Basic Data Access

One of the most important skills you will develop in this class is the ability to easily navigate datasets in Pandas. This is surprisingly tricky, so it has been broken down very slowly in this section. [Here is a useful guide.](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html)

#### Columns

Pandas DataFrames are essentially stored as a list of columns, which means that you don't need any fancy member functions to access them. If there are no spaces in the column name and it doesn't conflict with any pandas member function names, then you can access a column with a simple dot. For example, if you want to know the expression value for all genes in the base wild type glucose condition replicate 1, you access its column like this:

```python
data.base__wt_glc__1
```

If the column has a space or its name conflicts with a function name (e.g. "sum"), you have to use bracket notation. For example, there are spaces in the sample_table columns, so we would use this notation to see the carbon sources for all samples:

```python
sample_table['Carbon Source (g/L)']
```

You can also use bracket notation with a list to slice out multiple columns at once. Let's say you wanted all gene expression values, but only for the two base conditions:

```python
# make your list
cols_of_interest = ['base__wt_glc__1', 'base__wt_glc__2']

# index into the dataframe
data[cols_of_interest]
```

#### Rows

To get a row, you should use the "loc" function. Loc has many uses, which can get quite confusing. First, note that it can take a single row index or a list of row indices, just like the bracket notation for columns.

In the first cell, we'll get the expression of the gene 'b0002' across all samples.

In the second cell, we'll get the expression of three genes across all samples.

```python
data.loc['b0002']
```

```python
rows_of_interest = ['b0002', 'b0003', 'b0004']
data.loc[rows_of_interest]
```

#### Single Values and Slicing Both Ways at Once

To get a single value, you can either:

- Access the column first, then use bracket notation
- Use loc with a comma

Which one you use is a personal preference, but I tend to use the first where possible because it requires a bit less typing. As a beginner, it might be easier to stick with loc. 

Let's say you want to know how much the gene b0004 is expressed in the base condition, replicate 1:

```python
data.base__wt_glc__1['b0002']
```

```python
data.loc['b0002', 'base__wt_glc__1'] # row, column
```

Let's say you have a more specific slice you want. You can pass lists to both parts of loc if desired:

```python
data.loc[rows_of_interest, cols_of_interest]
```

#### Other Ways of Indexing

You may want to be familiar with [iloc](https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#indexing-integer), at, and iat. iloc, for example, allows you to index by number (as you would for lists and numpy arrays) instead of index names.

### Data Access: Using the Sample Table

Let's explore the information available to us in the sample table (for practice, you can try something similar with the gene table).

First, we can list the columns easily to know which features are known to us:

```python
sample_table.columns
```

#### Unique()

We can think of many of these columns as categorical variables, which have a finite set of available options. Let's use the [unique()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.unique.html) function to see which categories for Carbon Source exist in the dataset.

```python
sample_table['Carbon Source (g/L)'].unique()
```

So, each sample's Carbon Source may be one of: three concentrations of glucose (2, 4, and 5 g/L), two concentrations of acetate (2 and 3.5 g/L), or fumarate, pyruvate, or fructose. The last entry in the list, "nan", is short for "not a number", which is what Pandas uses to fill empty cells. In this case, it indicates that no carbon source was added to the base media. By the way, you can replace all nan's with another value using [fillna()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html?).

The unique function is useful for knowing which options are available, and it can also be used to iterate through all options, as we will see in later sections.

You may also want to know how many samples fall into each category; this can be done easily with the [value_counts()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.value_counts.html?highlight=value_counts#pandas.DataFrame.value_counts) function:

```python
sample_table['Carbon Source (g/L)'].value_counts()
```

#### Boolean Series

As shown above, there are 4 samples used fructose as a carbon source. Which samples are they, specifically? 

To find out, let's use the '==' operator to create a boolean (True/False) series for each sample. Note that the first equals sign is to assign this series to a variable, and the pair of equals signs compares each value in the column on the left to the value on the right.

```python
uses_fructose = sample_table['Carbon Source (g/L)'] == 'fructose(2)'
uses_fructose
```

In this series, each sample name is associated with a true/false value that indicates whether or not it uses fructose. We can use a series like this as a way of indexing! To get the names of the fructose-using samples, let's index into the row names themselves:

```python
fructose_samps = sample_table.index[uses_fructose]
fructose_samps
```

We can also use the boolean series with "loc" to get the rest of the sample table details about these samples:

```python
sample_table.loc[uses_fructose]
```

Lastly, let's say we want to get the gene expression data relating to these samples. In the data variable, samples are *columns*, even though they are rows in the sample table. The safest way to do this is to use the index names, instead of the boolean series.

```python
fru_data = data[fructose_samps]
fru_data
```

Let's repeat the above data access, but instead get the glucose(2) samples:

```python
uses_glu2 = sample_table['Carbon Source (g/L)'] == 'glucose(2)' # boolean series
glu2_samps = sample_table.index[uses_glu2] # sample names
glu2_data = data[glu2_samps] # slice of data
glu2_data
```

### Basic Statistics, Math, and Sorting: Glu v Fru Example

How does the choice of glucose versus fructose affect gene expression across all relevant conditions in the dataset?

We are already partway through answering this question, since we collected the data in section 5.1.2. Let's do a simple comparison of means - you'll explore more rigorous comparisons in the graded homeworks.

#### Results DataFrame and DataFrame Means

We'll start by computing the mean expression across *all columns* of both the fru_data and glu2_data. Think carefully about what this means: for each gene, we pool all conditions that have the given carbon source. 

To do this, use the [Pandas mean() function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html). By default, this function takes the mean down each column, but we want to take the mean across each row, so we pass the argument "axis = 1" to switch directions.

We'll save our results in a results DataFrame - this is a good way keep track of all your new computations in one place. Alternatively, you could have a bunch of separate Series. You wouldn't lose points for that approach, but it is a bit harder to keep it organized. 

Making the results DataFrame involves making a DataFrame nearly from scratch, which is covered in [this documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html). You don't need to populate it with values right away, but you should probably indicate what the indices and columns are from the get-go, as shown below.

```python
fru_data.mean(axis = 1)
```

```python
# set up your results dataframe
res_df = pd.DataFrame(index = data.index, # index is the same as data, i.e. one row for each gene
                      columns = ['glu2_avg', 'fru_avg']) # columns are the results (carbon source averages)

# fill in the glu2_avg column
res_df['glu2_avg'] = glu2_data.mean(axis = 1)
res_df['fru_avg'] = fru_data.mean(axis = 1)

# display the results
res_df
```

#### Subtraction and Absolute Value

As you can see above, some of the values are pretty close. It would be nice to know which genes are affected the most, which isn't clear yet. Let's get a metric for determining that - how about the difference between means?

Let's subtract one of our columns from the other, and make the result a new column in the dataframe:

```python
res_df['difference'] = res_df.fru_avg - res_df.glu2_avg

res_df
```

Since we want to know "which genes are affected most," we don't care if the genes are more highly expressed in glucose or in fructose. Therefore, we should take the [absolute value](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.abs.html). Again, save it as a new column:

```python
res_df['abs_diff'] = res_df.difference.abs()

res_df
```

#### Sort the DataFrame

Now, let's sort the results by the abs_diff column so that the genes that change the most go to the top. We'll use the [sort_values() function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sort_values.html?highlight=sort_values#pandas.DataFrame.sort_values) and pass "ascending = False" so that the largest differences show up first.

Here, we assign the result to the original DataFrame, replacing it in memory.

```python
res_df = res_df.sort_values('abs_diff', ascending = False)

res_df
```

The default pandas output above includes the 5 least changed genes, which is pretty useless. Let's use iloc to just grab the top 5 genes instead:

```python
top_genes = res_df.iloc[0:5]
top_genes
```

#### Connecting to Reality: Adding Gene Information

We have our top gene results above, so we technically know which genes react most strongly to a switch between glucose and fructose. We have "information", but no "knowledge". Let's add in some knowledge by finding the gene information relating to these genes:

```python
top_genes_info = gene_table.loc[top_genes.index]
top_genes_info
```

We can even combine the data and the gene details by [concatenating](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html) them together, which can be a useful way to summarize all results in a table:

```python
top_genes = pd.concat([top_genes, top_genes_info], axis = 1)
top_genes
```

These top genes make sense! Let's discuss why:

- the top gene, fruB, is strongly overexpressed in fructose. It is a fructose transporter, assisting in getting fructose into the cell. This would have a clear use in the fructose condition, which makes sense.
- the next genes, fruK and fruA, are in the same operon as fruB and are also important for fructose metabolism.
- the pyrBI genes are less obvious, but an [EcoCyc search](https://biocyc.org/gene?orgid=ECOLI&id=EG10805) shows that they are involved in the UMP Biosynthesis pathway. You could therefore predict that fructose affects nucleotide synthesis.

Connecting some knowledge to your results is a critical step; just finding top genes is useless unless you can gain some insight from them.
